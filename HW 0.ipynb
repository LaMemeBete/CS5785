```python
import matplotlib.pyplot as plt
from sklearn import datasets
import pandas as pd
import numpy as np
```

```python
iris_names=['sepal length in cm', 'sepal width in cm', 'petal length in cm', 'petal width in cm', 'class']
iris_data = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', names=iris_names)
```

```python
iris_data.describe()
iris_data['class'].unique()
print(iris_data.shape)
iris_data.head()
```

```python
#add color column
conditions  = [iris_data['class'] == 'Iris-setosa', iris_data['class'] == 'Iris-versicolor', iris_data['class'] == 'Iris-virginica']
choices     = ['red', 'blue', 'green']
iris_data['colors'] = np.select(conditions, choices)

column = list(iris_data)
fig, axes = plt.subplots(nrows=len(column)-2, ncols=len(column)-2, figsize=(20,20))
for i in range(len(column)):
    for j in range(len(column)):
        if((iris_data[column[i]].dtype == np.float64 or  iris_data[column[i]].dtype == np.int64) and (iris_data[column[j]].dtype == np.float64 or  iris_data[column[j]].dtype == np.int64)):
            '''if(i == j):
                fig.add_subplot.text(0.5, 0.5, 'title')
            else:'''
            iris_data.plot.scatter(column[i], column[j], c=iris_data['colors'], ax=axes[i,j])
```
